```
#Matías Ricardo Méndez
#A01422885
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import math

#Parametros del Grid
N = 22
grid_min = -10
grid_max = 10

#Carga (es importante definir cargas desde el principio)
q1= 20*np.exp(-6)
q1_loc = [0,0]

q2 = 20*np.exp(-6)
q2_loc = [5,5]

#Grids
x = np.linspace(grid_min,grid_max,N)
y = np.linspace(grid_min,grid_max,N)

x1 = x-q1_loc[0]
y1= y-q1_loc[1]

x2 = x-q2_loc[0]
y2 = y-q2_loc[1]

#Matrices
X1,Y1 = np.meshgrid(x1,y1)
X2,Y2 = np.meshgrid(x2,y2)

#Constante K
eps_0 = 8.8542*np.exp(-12)
k_e = 1/(4*np.pi*eps_0)

#Angulos

angle1 = np.angle(X1+(Y1*(1j)))

angle2 = np.angle(X2+(Y2*(1j)))

#Radio
r1 = np.sqrt((x1**2)+(y1**2))

r2 = np.sqrt((x2**2)+(y2**2))

#Componentes en ambos ejes
Ex1 = k_e*(q1/(r1**2))*np.cos(angle1)
Ey1 = k_e*(q1/(r1**2))*np.sin(angle1)

Ex2 = k_e*(q2/(r2**2))*np.cos(angle2)
Ey2 = k_e*(q2/(r2**2))*np.sin(angle2)

#Magnitud de los componetes usando pitagoras
mags1 = np.sqrt((Ex1**2)+(Ey1**2))

mags2 = np.sqrt((Ex2**2)+(Ey2**2))

#Componentes Unitarios (compenentes en cada / magnitud), esto nos permitirá que todas las felchas sean iguales
Ex_unit1 = Ex1/mags1
Ey_unit1 = Ey1/mags1

Ex_unit2 = Ex2/mags2
Ey_unit2 = Ey2/mags2


# Creamos una figura y eje con plt.subplots()
fig, ax = plt.subplots(figsize = (7,7))

# Agregamos al eje las flechas de los vectores con plt.quiver()
#ax.quiver(X1,Y1,Ex_unit1,Ey_unit1)
#ax.quiver(X2,Y2,Ex_unit2,Ey_unit2)

# Dibujamos la carga puntual usando plt.scatter()
ax.scatter(q1_loc[0],q1_loc[1],c='red',s=500)
ax.scatter(q2_loc[0],q2_loc[1],c='black',s=500)

#Configuramos las dimensiones del eje y el aspecto
ax.axis([grid_min,grid_max,grid_min,grid_max])
ax.set_aspect('equal','box')

#Paso 3, obtener Ex y Ey totales de ambas cargas y Graficarlos
#Sacar totales
ExT = Ex1 + Ex2
EyT = Ey1 + Ey2

#Sacaré magnitudes a ver si funciona para asi poder sacar componentes unitarios y eso graficarlo
#Magnitudes totales
magsT = np.sqrt((ExT**2)+(EyT**2))

#Componentes unitarios totales
ExT_unit = ExT/magsT
EyT_unit = EyT/magsT

#Finalmente se gráfican los vectores unitarios para que las flechas sean del mismo tamaño y se usa el rango y dominio de X1 y Y1

ax.quiver((X1),(Y1),ExT_unit,EyT_unit)

```
